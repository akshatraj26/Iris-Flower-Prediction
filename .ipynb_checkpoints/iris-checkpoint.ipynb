## A box and whisker plot—also called a box plot—displays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum. # 9. Violin Plot : It is used to visualize the distribution of data and its probability distribution. This chart is a combination of a boxplot and a Density Plot that is rotated and placed on each side, to show the distribution shape of the data. The thick black bar in the centre represents the interquartile range,  the thin black line extended from it represents the 95% confidence intervals, and the white dot is the median.  Box plots are limited in their  display of the data, as their visual simplicity tends to hide significant details about how values about how values in the data are distributed.





## Black thin line in the centre is interquartile.

## White dot is a median.

```python
fig = plt.gcf()
fig.set_size_inches(10, 7)
fig  = sns.violinplot(x ='Species', y = 'PetalLengthCm', data =iris)
```

<h3 style="color:purple">Categorical Endoding</h3>

```python
iris = iris.drop('Id', axis=1)
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
flower_class = {'Iris-setosa' : 0,
                'Iris-versicolor' : 1,
                'Iris-virginica' : 2}

iris['Species'] = iris['Species'].map(flower_class)
```

```python
iris.head()
```

<h3 style="color:purple">Dependent Variable and Independent Variable</h3>

```python
iris.columns
```

```python
X = iris.drop('Species', axis =1)
y = iris['Species']
```

<h3 style="color:purple">Split the data into train test and Split</h3>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<h3 style="color:purple">Build the Model</h3>

```python
from xgboost import XGBClassifier
xg = XGBClassifier()
```

```python
xg.fit(X_train, y_train)
```

```python
pred = xg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
cm = confusion_matrix(y_test, pred)
cm
```

<h3 style="color:purple">Heat Map</h3>

```python
sns.heatmap(cm,  annot =True, cmap = 'Blues')
plt.show()
```

<h3 style="color:purple">Accuracy Score</h3>

```python
ac = accuracy_score(y_test, pred)
ac
```

<h3 style="color:purple">Save the Model</h3>

```python
import joblib
joblib.dump(xg, 'flower_model.joblib')
```
